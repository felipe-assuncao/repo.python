#### Questão 12
Um estabelecimento está vendendo combustíveis com descontos variados. Para o etanol, se a quantidade comprada for até 15 litros, o desconto será de 2% por litro. Caso contrário, será de 4% por litro. Para o diesel, se a quantidade comprada for até 15 litros, o desconto será de 3% por litro. Caso contrário, será de 5% por litro. O preço do litro de diesel é R\$ 2,00 e o preço do litro de etanol é R\$ 1,70. Escreva um programa que leia a quantidade de litros vendidos e o tipo de combustível (E para etanol e D para diesel) e calcule o valor a ser pago pelo cliente. Tenha em mente algumas dicas:

- O do valor do desconto será a multiplicação entre preço do litro, quantidade de litros e o valor do desconto.
- O valor a ser pago por um cliente será o resultado da multiplicação do preço do litro pela quantidade de litros menos o valor de desconto resultante do cálculo.


````python
litros = float(input("Quantidade de litros: "))
tipo = input("Tipo de combustível (E-Etanol, D-Diesel): ").strip().upper()

# Preços definidos pela questão
preco_diesel = 2.00
preco_etanol = 1.70

if tipo == "E":
    preco_litro = preco_etanol
    if litros <= 15:
        percentual_desconto = 0.02  # 2%
    else:
        percentual_desconto = 0.04  # 4%
        
elif tipo == "D":
    preco_litro = preco_diesel
    if litros <= 15:
        percentual_desconto = 0.03  # 3%
    else:
        percentual_desconto = 0.05  # 5%
else:
    print("Tipo inválido!")
    percentual_desconto = -1 # Sinalizador de erro

if percentual_desconto != -1:
    # 1. "O valor do desconto será a multiplicação entre preço do litro, quantidade de litros e o valor do desconto"
    valor_desconto = preco_litro * litros * percentual_desconto
    
    # 2. "O valor a ser pago será (preço * litros) menos o valor de desconto"
    valor_total_bruto = preco_litro * litros
    valor_pago = valor_total_bruto - valor_desconto
    
    print(f"Valor a pagar: R$ {valor_pago:.2f}")






````

#### Questão 7
Os números primos possuem várias aplicações dentro da Ciência de Dados, por exemplo, na criptografia e segurança. Um número primo é aquele que é divisível apenas por um e por ele mesmo. Faça um programa que peça um número inteiro e determine se ele é ou não um número primo.

````python
numero = int(input('Digíte um número inteiro.'))
e_primo = True

if numero < 2:
    e_primo = False
else:
    for i in range(2, numero):
        if numero % i == 0:
            e_primo = False
            break

if e_primo:
    print(f'O número {numero} é primo!')
else:
    print(f'O número {numero} não é primo!')
    

    
````

Dica: para calcular o percentual de crescimento usamos a seguinte equação: `100 * (amostra_atual - amostra_passada) / (amostra_passada)`.

````python
lista = [1.2, 2.1, 3.3, 5.0, 7.8, 11.3, 16.6, 25.1, 37.8, 56.9]
lista_percentual = []
for i in range(1, len(lista)):
    amostra_atual = lista[i]
    amostra_passada = lista[i - 1]
    percentual = (100 * (amostra_atual - amostra_passada) / (amostra_passada)) 
    lista_percentual.append(round(percentual,1))
print (lista_percentual)
````

#### Questão 13
Os funcionários de um setor da empresa que você trabalha vão receber um abono correspondente a 10% do seu salário devido ao ótimo desempenho do time. O setor financeiro solicitou sua ajuda para a verificação das consequências financeiras que esse abono irá gerar nos recursos. Assim, foi encaminhada para você uma lista com os salários que receberão o abono: `[1172, 1644, 2617, 5130, 5532, 6341, 6650, 7238, 7685, 7782, 7903]`. O abono de cada funcionário não pode ser inferior a 200. Em código, transforme cada um dos salários em chaves de um dicionário e o abono de cada salário no elemento. Depois, informe o total de gastos com o abono, quantos funcionários receberam o abono mínimo e qual o maior valor de abono fornecido.


````python
salarios = [1172, 1644, 2617, 5130, 5532, 6341, 6650, 7238, 7685, 7782, 7903]
abonos_dict = {}
total_gastos = 0
funcionarios_minimo = 0
valor_minimo = 200.0

# 1. Transformar salários em chaves e calcular abonos
for salario in salarios:
    # Calcula 10%, mas garante no mínimo R$ 200.00
    abono = max(salario * 0.10, valor_minimo)
    abonos_dict[salario] = round(abono, 2)
    
    # Acumula total para o financeiro
    total_gastos += abonos_dict[salario]
    
    # Conta quantos receberam o mínimo
    if abono == valor_minimo:
        funcionarios_minimo += 1

# 2. Informações solicitadas
maior_abono = max(abonos_dict.values())

print("Dicionário de Abonos (Salário: Abono):")
print(abonos_dict)
print("-" * 30)
print(f"Total de gastos com abono: R$ {total_gastos:.2f}")
print(f"Funcionários com abono mínimo (R$ 200): {funcionarios_minimo}")
print(f"Maior valor de abono fornecido: R$ {maior_abono:.2f}")
````

#### Questão 15
O setor de RH da sua empresa te pediu uma ajuda para analisar as idades dos funcionários de 4 setores da empresa. Para isso, ele te forneceu os seguintes dados:
```
{'Setor A': [22, 26, 30, 30, 35, 38, 40, 56, 57, 65],
 'Setor B': [22, 24, 26, 33, 41, 49, 50, 54, 60, 64],
 'Setor C': [23, 26, 26, 29, 34, 35, 36, 41, 52, 56],
 'Setor D': [19, 20, 25, 27, 34, 39, 42, 44, 50, 65]}
``` 
Sabendo que cada setor tem 10 funcionários, construa um código que calcule a média de idade de cada setor, a idade média geral entre todos os setores e quantas pessoas estão acima da idade média geral.


````python
# Dados fornecidos pelo RH
dados_setores = {
    'Setor A': [22, 26, 30, 30, 35, 38, 40, 56, 57, 65],
    'Setor B': [22, 24, 26, 33, 41, 49, 50, 54, 60, 64],
    'Setor C': [23, 26, 26, 29, 34, 35, 36, 41, 52, 56],
    'Setor D': [19, 20, 25, 27, 34, 39, 42, 44, 50, 65]
}

# 1. Média de idade de cada setor
total_idades_geral = 0
total_funcionarios = 0

for setor, idades in dados_setores.items():
    media_setor = sum(idades) / len(idades)
    print(f"Média de idade do {setor}: {media_setor:.1f} anos")
    
    total_idades_geral += sum(idades)
    total_funcionarios += len(idades)

# 2. Idade média geral
media_geral = total_idades_geral / total_funcionarios
print(f"\nIdade média geral de todos os setores: {media_geral:.1f} anos")

# 3. Quantas pessoas estão acima da idade média geral
pessoas_acima_media = 0
for idades in dados_setores.values():
    for idade in idades:
        if idade > media_geral:
            pessoas_acima_media += 1

print(f"Quantidade de pessoas acima da média geral: {pessoas_acima_media}")
````
